# Paris Metropolitan, an evolution
The goal of this notebook is to compare the database in xlsx format with the updated geojson file and to ouput a corresponding csv file that includes new information added

```python
#imports
import pandas as pd
import numpy as np
import json
```

```python
#path to database
json_database_path = r'paris_metropolitan_evolution_db_updated.geojson'
```

```python
#open xlsx file and json file
paris_metro_xls_db = pd.read_excel('paris_metropolitan_evolution_db.xlsx', skiprows=1)
paris_metro_json_db = json.load(open(json_database_path))
```

```python
#iterate through both files in order to update the first database in xlsx format with the second one
for index, row in paris_metro_xls_db.iterrows() :
    for feature in paris_metro_json_db['features'] :
        if row['stop_id'] == feature['properties']['stop_id'] :
            paris_metro_xls_db.loc[index,'stop_info'] = feature['properties']['stop_info']
            paris_metro_xls_db.loc[index,'open_date'] = feature['properties']['open_date']
            paris_metro_xls_db.loc[index,'close_date'] = feature['properties']['close_date']
            paris_metro_xls_db.loc[index, 'ratp_id'] = feature['properties']['ratp_id']
            break
paris_metro_xls_db
```

```python
#write a csv file based on the updated xlsx database
file_name = "final_database_files/paris_metropolitan_evolution_db.csv"
paris_metro_xls_db.to_csv(file_name, sep='\t', encoding='utf-8')
```
